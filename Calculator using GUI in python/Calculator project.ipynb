Calculator

```python
from tkinter import*
root=Tk()
root.title('Simple calculator')

e=Entry(root,width=35,borderwidth=5)
e.grid(row=0,column=0,columnspan=3,padx=10,pady=10)

def button_click(n):
    cur=e.get()
    e.delete(0, END)
    e.insert(0, cur+n)
    
def button_clear():
    e.delete(0, END)
    
def button_add():
    global f_num
    f_num=int(e.get())
    e.delete(0,END)
    global temp
    temp="+"

def button_substract():
    global f_num
    f_num=int(e.get())
    e.delete(0,END)
    global temp
    temp='-'
    
def button_multiply():
    global f_num
    f_num=int(e.get())
    e.delete(0,END)
    global temp
    temp="*"
    

def button_divide():
    global f_num
    f_num=int(e.get())
    e.delete(0,END)
    global temp
    temp='/'
    
def button_equal():
    suc=int(e.get())
    e.delete(0,END)
    
    if temp == '+':
        e.insert(0,suc+f_num)
    elif temp == '-':
        e.insert(0,f_num-suc)
    elif temp == '*':
        e.insert(0,f_num*suc)
    elif temp == '/':
        e.insert(0,f_num/suc)
        


#define button
 
button_1=Button(root,text="1",padx=30,pady=10,command = lambda:button_click("1"))
button_2=Button(root,text="2",padx=30,pady=10,command = lambda:button_click("2"))
button_3=Button(root,text="3",padx=30,pady=10,command = lambda:button_click("3"))
button_4=Button(root,text="4",padx=30,pady=10,command = lambda:button_click("4"))
button_5=Button(root,text="5",padx=30,pady=10,command = lambda:button_click("5"))
button_6=Button(root,text="6",padx=30,pady=10,command = lambda:button_click("6"))
button_7=Button(root,text="7",padx=30,pady=10,command = lambda:button_click("7"))
button_8=Button(root,text="8",padx=30,pady=10,command = lambda:button_click("8"))
button_9=Button(root,text="9",padx=30,pady=10,command = lambda:button_click("9"))
button_0=Button(root,text="0",padx=29,pady=10,command = lambda:button_click("0"))

button_clear=Button(root,text="Clear",padx=60,pady=10,command=button_clear)
button_add=Button(root,text="+",padx=29,pady=10,command= button_add)
button_equal=Button(root,text="=",padx=67,pady=10,command=button_equal)

button_multiply=Button(root,text="*",padx=30,pady=10,command=button_multiply)
button_divide=Button(root,text='/',padx=30,pady=10,command=button_divide)
button_substract=Button(root,text="-",padx=30,pady=10,command=button_substract)


#putting it to screen

button_1.grid(row=3,column=0)
button_2.grid(row=3,column=1)
button_3.grid(row=3,column=2)

button_4.grid(row=2,column=0)
button_5.grid(row=2,column=1)
button_6.grid(row=2,column=2)

button_7.grid(row=1,column=0)
button_8.grid(row=1,column=1)
button_9.grid(row=1,column=2)

button_0.grid(row=4,column=0)
button_clear.grid(row=4,column=1,columnspan=2)
                     
button_add.grid(row=5,column=0)
button_equal.grid(row=5,column=1,columnspan=2)
                     
button_multiply.grid(row=6,column=0)
button_divide.grid(row=6,column=1)     
button_substract.grid(row=6,column=2)

root.mainloop()
```
